```python
# Importing libraries and date

import pandas as pd
pd.set_option("display.max_columns",None)
df = pd.read_csv("Airbnb_Open_Data.csv")
import numpy as np
```

```python
# Take a samole of a data

df.sample(10)
```

```python
df.info()
```

# The columns i will modify on are :-

- "host_identity_verified"
- "country"
- "price"
- "service fee"
- "last review"

```python
df["host_identity_verified"].unique()
```

```python
# Changing the column "host_identity_verified" values

df["host_identity_verified"] = df["host_identity_verified"].str.replace("unconfirmed","NO")
df["host_identity_verified"] = df["host_identity_verified"].str.replace("verified","YES")
df["host_identity_verified"] = df["host_identity_verified"].fillna("")
```

```python
# Drop country "column" cause it is useless we have "counrty code" column

df = df.drop(columns="country")
```

```python
df["Construction year"].unique()
```

```python
# Changing the type of the "Construction year" column to be numerical

df['Construction year'] = df['Construction year'].replace([np.nan, np.inf, -np.inf], np.nan).fillna(0).astype(int)
```

```python
df['Construction year'].unique()
```

```python
df["price"].unique()
```

```python
# Changing the type of the "price" column to be numerical


df["price"] = df["price"].str.replace("$","")
df["price"] = df["price"].str.replace(",","")
df["price"] = df["price"].astype(float)
```

```python
df["price"].unique()
```

```python
df["service fee"].unique()
```

```python
# Changing the type of the "service fee" column to be numerical

df["service fee"] = df["service fee"].str.replace("$","")
df["service fee"] = df["service fee"].astype(float)
```

```python
df["service fee"].unique()
```

```python
# Changing the type of the "last review" column to be date

df['last review'] = pd.to_datetime(df['last review'])
```

```python
# Changing the name of the columns to be the first letter upper

df.columns = df.columns.str.title()
```
